```python
import math


def exercise1 ( tp , fp , fn ) :
    if type(tp) is not int:
        print('tp must be int')
        return
    if type(fp) is not int:
        print('fp must be int')
        return
    if type(fn) is not int:
        print('fn must be int')
        return
    if tp <= 0 or fp <= 0 or fn <= 0:
        print('tp and fp and fn must be greater than zero')
        return
    precision = tp / (tp + fp)
    recall = tp / (tp + fn)
    f1_score = 2 * (precision * recall) / (precision + recall)
    
    print('Precision:', precision)
    print('Recall:', recall)
    print('F1-score:', f1_score)
    
tp = int(input('Input tp'))    
fp = int(input('Input fp'))  
fn = int(input('Input fn'))  
exercise1(tp, fp, fn)
```

```python
import math

def is_number(n):
    try:
        float(n)
        return True
    except ValueError:
        return False

def sigmoid(x):
    return 1 / (1 + math.exp(-x))

def relu(x):
    return max(0, x)

def elu(x, alpha=0.01):
    if x > 0:
        return x 
    else:
        result = alpha * (math.exp(x) - 1)
        return result

def exercise2():
    x = input("Input x: ")
    if not is_number(x):
        print('x must be a number')
        return
    x = float(x)
    activation_function = input("Input activation function (sigmoid | relu | elu): ").lower()
    if activation_function not in ['sigmoid', 'relu', 'elu']:
        print(f'{activation_function} is not supported')
        return
    if activation_function == 'sigmoid':
        result = sigmoid(x)
    elif activation_function == 'relu':
        result = relu(x)
    elif activation_function == 'elu':
        result = elu(x)
    print(f'{activation_function}: f({x}) = {result}')
exercise2()

```

```python
import random
import math

def is_number(n):
    try:
        float(n)
        return True
    except ValueError:
        return False

def mae_loss(y, y_hat, num_samples):
    total_loss = 0
    for i in range(num_samples):
        total_loss += abs(y[i] - y_hat[i])
    return total_loss / num_samples

def mse_loss(y, y_hat, num_samples):
    total_loss = 0
    for i in range(num_samples):
        total_loss += (y[i] - y_hat[i]) ** 2
    return total_loss / num_samples

def rmse_loss(y, y_hat, num_samples):
    return math.sqrt(mse_loss(y, y_hat, num_samples))

def exercise3():
    num_samples = input("Input number of samples (integer number) which are generated: ")
    if not is_number(num_samples):
        print('number of samples must be an integer number')
        return
    num_samples = int(num_samples)
    loss_name = input("Input loss name (MAE, MSE, RMSE): ").upper()
    y = [random.uniform(0, 10) for _ in range(num_samples)]
    y_hat = [random.uniform(0, 10) for _ in range(num_samples)]
    if loss_name == 'MAE':
        loss = mae_loss(y, y_hat, num_samples)
    elif loss_name == 'MSE':
        loss = mse_loss(y, y_hat, num_samples)
    elif loss_name == 'RMSE':
        loss = rmse_loss(y, y_hat, num_samples)
    for i in range(num_samples):
        if loss_name == 'MAE':
            print(f'loss name: {loss_name}, sample: {i}, y_hat: {y_hat[i]}, y: {y[i]}, loss: {abs(y_hat[i] - y[i])}')
        else:
            print(f'loss name: {loss_name}, sample: {i}, y_hat: {y_hat[i]}, y: {y[i]}, loss: {(y_hat[i] - y[i]) ** 2}')        
    print(f'final {loss_name}: {loss}')

exercise3()
```

```python
import math

def is_number(n):
    try:
        float(n)
        return True
    except ValueError:
        return False

def factorial(n):
    if n == 0:
        return 1
    else:
        return n * factorial(n - 1)

def sin(x, n):
    result = 0
    for i in range(n):
        term = ((-1) ** i) * (x ** (2 * i + 1)) / factorial(2 * i + 1)
        result += term
    return result

def cos(x, n):
    result = 0
    for i in range(n):
        term = ((-1) ** i) * (x ** (2 * i)) / factorial(2 * i)
        result += term
    return result

def sinh(x, n):
    result = 0
    for i in range(n):
        term = (x ** (2 * i + 1)) / factorial(2 * i + 1)
        result += term
    return result

def cosh(x, n):
    result = 0
    for i in range(n):
        term = (x ** (2 * i)) / factorial(2 * i)
        result += term
    return result

def exercise4():
    n = int(input("Input n: "))
    x = input("Input x: ")
    if not is_number(x):
        print('x must be a number')
        return
    x = float(x)
    operation = input("Input operation (sin | cos | sinh | cosh): ").lower()
    if operation not in ['sin', 'cos', 'sinh', 'cosh']:
        print(f'{operation} is not supported')
        return
    if operation == "sin":
        result = sin(x, n)
    elif operation == "cos":
        result = cos(x, n)
    elif operation == "sinh":
        result = sinh(x, n)
    elif operation == "cosh":
        result = cosh(x, n)
    print(f"{operation}(x = {x}, n = {n})")
    print(f" {result}")
exercise4()
```
